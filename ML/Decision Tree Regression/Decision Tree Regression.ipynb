```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
dataset = pd.read_csv('50_Startups.csv')
```

```python
dataset
```

```python
modified_dataset = pd.get_dummies(dataset, drop_first=True)
```

```python
modified_dataset
```

```python
inp_var = modified_dataset[['R&D Spend','Administration', 'Marketing Spend','State_Florida', 'State_New York']]
out_var = modified_dataset[['Profit']]
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(inp_var, out_var, test_size = 1/3, random_state=0)
```

```python
x_train.shape
```

```python
x_test.shape
```

```python
y_train.shape
```

```python
y_test.shape
```

```python
from sklearn.tree import DecisionTreeRegressor
regressor = DecisionTreeRegressor(criterion='friedman_mse')
regressor = regressor.fit(x_train, y_train)
```

```python
y_pred = regressor.predict(x_test)
```

```python
from sklearn.metrics import r2_score
r_sq = r2_score(y_pred, y_test)
```
