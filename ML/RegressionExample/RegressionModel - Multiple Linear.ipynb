```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
dataset = pd.read_csv('insurance_pre.csv')
```

```python
dataset
```

```python
dataset.columns
```

```python
modified_dataset = pd.get_dummies(dataset, drop_first = True)
```

```python
modified_dataset
```

```python
indep_var = modified_dataset[['age', 'bmi', 'children','sex_male','smoker_yes']]
dep_var = modified_dataset[['charges']]
```

```python
indep_var
```

```python
dep_var
```

```python
# Training and test data split
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(indep_var, dep_var, test_size = 1/3, random_state = 0)
```

```python
x_train
```

```python
x_test
```

```python
x_train.shape
```

```python
x_test.shape
```

```python
y_train
```

```python
y_test
```

```python
y_train.shape
```

```python
y_test.shape
```

```python
from sklearn.linear_model import LinearRegression
regressor = LinearRegression(fit_intercept=False)
regressor.fit(x_train, y_train)
```

```python
weight=regressor.coef_
print("Weight of the model={}".format(weight))
bais=regressor.intercept_
print("Intercept of the model={}".format(bais))
```

```python
weight
```

```python
bais
```

```python
y_pred = regressor.predict(x_test)
```

```python
from sklearn.metrics import r2_score
r_score=r2_score(y_test,y_pred)
```
